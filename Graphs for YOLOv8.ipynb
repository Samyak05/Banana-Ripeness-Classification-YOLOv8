```python
import numpy as np
import matplotlib.pyplot as plt

# Training loss values (example data)
epochs = list(range(1, 51))
loss_values = [
    0.5558, 0.2094, 0.1854, 0.1668, 0.1391,
    0.1175, 0.11, 0.1049, 0.09017, 0.09451,
    0.08358, 0.07749, 0.07202, 0.07246, 0.01998
]

# Linear interpolation to fill missing loss values
missing_epochs = np.arange(1, 51)
interpolated_loss_values = np.interp(missing_epochs, epochs[:len(loss_values)], loss_values)

# Set the last epoch's loss value
interpolated_loss_values[-1] = loss_values[-1]

# Create a line plot for training loss
plt.figure(figsize=(8, 6))
plt.plot(missing_epochs, interpolated_loss_values, marker='o', color='b', label='Training Loss')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Training loss data (example values, replace with your actual data)
epochs = list(range(1, 51))
training_loss = [0.5558, 0.2094, 0.1854, 0.1668, 0.1391, 0.1175, 0.11, 0.1049, 0.09017, 0.09451,
                 0.08358, 0.07749, 0.07202, 0.07246]

# Validation loss data (example values, replace with your actual data)
validation_loss = [0.01998, 0.01812, 0.01673, 0.01521, 0.01405, 0.01309, 0.01234, 0.01172, 0.01092,
                  0.01067, 0.01012, 0.00978, 0.00942, 0.00931]

# Linear interpolation to fill missing training loss values
missing_epochs = np.arange(1, 51)
interpolated_training_loss = np.interp(missing_epochs, epochs[:len(training_loss)], training_loss)

# Set the last training epoch's loss value
interpolated_training_loss[-1] = training_loss[-1]

# Linear interpolation to fill missing validation loss values
interpolated_validation_loss = np.interp(missing_epochs, epochs[:len(validation_loss)], validation_loss)

# Set the last validation epoch's loss value
interpolated_validation_loss[-1] = validation_loss[-1]

# Create a line plot for training and validation loss
plt.figure(figsize=(8, 6))
plt.plot(missing_epochs, interpolated_training_loss, marker='o', color='b', label='Training Loss')
plt.plot(missing_epochs, interpolated_validation_loss, marker='x', color='r', label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.grid(True)
plt.legend()
plt.show()
```
